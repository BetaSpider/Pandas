```python
def major(a): #create a funmction called 'major'
    if a>17: #if the age is greater than 17, create a new calumn called 'Eligible_drinker' & set it as True
        return True
    else:
        return False
    
df2['Eligible_drinker']=df2['age'].apply(major)
df2.head()
df2.tail()
```

```python
def times(x): #create a function called times
    if type(x) is int: if the data type of the parameter is int, then multiply it by 10 and return the same.
        return 10 * x
    return x
df2.applymap(times).head()
```
